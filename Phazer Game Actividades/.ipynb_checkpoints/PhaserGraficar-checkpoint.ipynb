```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(r'C:\Users\rovega\Desktop\ITM\Sem 9\Inteligencia Artificial\Git IA - Actividades\Phazer Game Actividades\CSV-Datos-Phaser\game_data_1.csv', header=None, names=['Distance', 'Speed', 'PlayerStatus'], dtype=float)

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points for target = 0 in blue
ax.scatter(df[df['PlayerStatus'] == 0]['Distance'], df[df['PlayerStatus'] == 0]['Speed'], [0] * len(df[df['PlayerStatus'] == 0]['Distance']), c='blue', marker='o', label='PlayerStatus=0')

# Plot the data points for target = 1 in red
ax.scatter(df[df['PlayerStatus'] == 1]['Distance'], df[df['PlayerStatus'] == 1]['Speed'], [1] * len(df[df['PlayerStatus'] == 1]['Distance']), c='red', marker='x', label='PlayerStatus=1')

# Set labels for the axes
ax.set_xlabel('Distance')
ax.set_ylabel('Speed')
ax.set_zlabel('PlayerStatus')

# Add a legend
ax.legend()

# Show the plot
plt.show()

```
